```python
import pyspark
from motordecalidad.functions import *
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Motor de Calidad").getOrCreate()
```

```python
df = spark.read.option("header",True).csv('MOCK_DATA.csv')
```

```python
Codigo,Ratio,NullAmnt = validateNull(df,"ID_CLIENTE")
```

```python
Codigo2, Ratio2, DuplicatedAmount = validateDuplicates(df,["ID_CLIENTE"])
```

```python
OutputDataFrameColumns = ["CODIGO_DE_REGLA","RATIO_DE_ERROR","CANTIDAD_DE_REGISTROS_CON_ERROR"]
Data = [(Codigo,Ratio,NullAmnt),(Codigo2,Ratio2,DuplicatedAmount)]
```

```python
OutputDataFrame = spark.createDataFrame(data = Data, schema = OutputDataFrameColumns)
```
